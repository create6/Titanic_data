# 绘制电影票房的柱状图

```python
# 1. 准备数据
names = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴', '降魔传','追捕','七十七天','密战','狂兽','其它']
income = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]
# 2. 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 3. 绘制柱状图
x = range(len(names))
plt.bar(x, income, width=0.5, color=['r','g','b','c','m'])

# 4. 设置x轴刻度
plt.xticks(x, names)

# 4. 显示
plt.show()
```
