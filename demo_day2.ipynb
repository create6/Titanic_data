### 1.Numpy基本操作


```python
import matplotlib.pyplot as plt
import numpy as np
#准备数据
x = np.linspace(-10,10,1000)
# y_sin = np.sin(x)
y_cos = np.cos(x)
#创建画布
plt.figure(figsize=(20,8),dpi=100)
# plt.plot(x,y_sin)
plt.plot(x,y_cos)
#绘图
plt.xlabel("x",fontsize=25)  #没有等号.TypeError: 'str' object is not callable
plt.ylabel("y",fontsize=25)
plt.title("cos函数图形",fontsize=25)
# 添加网格
plt.grid(True, linestyle='--', alpha=0.5, color='r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5), dpi=80)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="Cos Function")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="Sin Function")

plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])

plt.yticks([-1, +1])

t = 2 * np.pi / 3
plt.plot([t, t], [0, np.cos(t)],
         color='blue', linewidth=1.5, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')


plt.plot([t, t], [0, np.sin(t)],
         color='red', linewidth=1.5, linestyle="--")
plt.scatter([t, ], [np.sin(t), ], 50, color='red')


plt.legend(loc='upper left')
plt.savefig('aaa.png')
plt.show()
```

##### 归一化

```python
# 请将以下数据归一化到0~1之间   说明: 归一化公式为: (x-min)/(max-min) , 
# 其中min该特征的最小值, max为该特征的最大值, x为具体的特征值
import numpy as np
#准备数据
data = np.array([
    [90,2,10,40],
    [60,4,15,45],
    [75,3,13,46]
])
#计算每一列的最大值
max_value=np.max(data,axis=0)
#计算每一列的最小值
min_value = np.min(data,axis=0)
#4,根据公式计算,进行归一化处理
new_data = (data -min_value)/(max_value-min_value)
#打印归一化后的数据
print(new_data)


```
