# 保存图片

```python
# 1. 导包
import matplotlib.pyplot as plt
# 2. 创建画布
plt.figure(figsize=(20, 10), dpi=100)
# 3. 准备数据
x = [1, 2, 3, 4, 5, 6, 7]
y = [10,15,12,16,11,18,9]
# 4. 画图
plt.plot(x, y)

# 保存图片
plt.savefig('test.png')
# 注意: 保存图片必须在show之前. 
# 5. 展示
plt.show()
```

# 绘制温度变化情况

```python
import matplotlib.pyplot as plt
import random
```

```python
# 1. 准备数据
x = range(60)
y = [random.uniform(30, 38) for i in x]
# 2. 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 3. 绘制折线图
plt.plot(x, y)


# 添加刻度
# 准备x轴刻度标签
x_tick_lables = ['11点{}分'.format(i) for i in x]
# print(x_tick_lables)
# y轴的刻度
y_ticks = range(40)

# 添加x轴刻度
plt.xticks(x[::5], x_tick_lables[::5], fontsize=20)
# 添加y轴刻度
plt.yticks(y_ticks[::5], fontsize=20)


# 添加网格
plt.grid(True, linestyle='--', alpha=0.5, color='r')

# 添加描述信息
# 添加x轴名称
plt.xlabel('时间', fontsize=22)
# 添加y轴名称
plt.ylabel('温度', fontsize=22)
# 添加标题
plt.title('广州11点0分到12点的温度变化情况', fontsize=24)


# 4. 展示
plt.show()
```

## 多次plot

```python
# 1. 准备数据
x = range(60)
y_gz = [random.uniform(30, 38) for i in x]
# 北京温度
y_bj = [random.uniform(1, 5) for i in x]

# 2. 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 3. 绘制折线图
plt.plot(x, y_gz, color='r', linestyle='--', label='广州')
plt.plot(x, y_bj, color='b', label='北京')

# 设置图例
plt.legend(loc='best')

# 添加刻度
# 准备x轴刻度标签
x_tick_lables = ['11点{}分'.format(i) for i in x]
# print(x_tick_lables)
# y轴的刻度
y_ticks = range(40)

# 添加x轴刻度
plt.xticks(x[::5], x_tick_lables[::5], fontsize=20)
# 添加y轴刻度
plt.yticks(y_ticks[::5], fontsize=20)


# 添加网格
plt.grid(True, linestyle='--', alpha=0.5, color='r')

# 添加描述信息
# 添加x轴名称
plt.xlabel('时间', fontsize=22)
# 添加y轴名称
plt.ylabel('温度', fontsize=22)
# 添加标题
plt.title('广州11点0分到12点的温度变化情况', fontsize=24)


# 4. 展示
plt.show()
```

# 多坐标系

```python
import random
import matplotlib.pyplot as plt
# 1. 准备数据
x = range(60)
y_gz = [random.uniform(30, 38) for i in x]
# 北京温度
y_bj = [random.uniform(1, 5) for i in x]

# 2. 获取坐标系
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=100)

# 3. 绘制折线图
axes[0].plot(x, y_gz, linestyle='--', color='r', label='广州')
axes[1].plot(x, y_bj, color='b', label='北京')

# 4. 添加描述信息
# 4.1 添加刻度
x_tick_labels = ['11点{}分'.format(i) for i in x]
y = range(40)
y_tick_labels = ['{}℃'.format(i) for i in y]
# axes[0].set_xticks(x[::10])
# axes[0].set_xticklabels(x_tick_labels[::10])
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_tick_labels[::5])

axes[0].set_yticks(y[::5])
axes[0].set_yticklabels(y_tick_labels[::5])

axes[1].set_xticks(x[::10])
axes[1].set_xticklabels(x_tick_labels[::10])
axes[1].set_yticks(y[::5])

# 设置坐标轴名称
axes[0].set_xlabel('时间')
axes[0].set_ylabel('温度')
axes[0].set_title('广州的温度变化情况')

axes[1].set_xlabel('时间')
axes[1].set_ylabel('温度')
axes[1].set_title('北京的温度变化情况')

# 图例
axes[0].legend()
axes[1].legend()

# 网格
axes[0].grid(linestyle='--', color='g', alpha=0.5)
axes[1].grid(linestyle='-', color='r', alpha=0.5)

# 5. 显示
plt.show()
```

# 折线图的应用场景

```python
# 绘制sin()图像
import numpy as np

# 1.准备数据
x = np.linspace(-10, 10, 1000)
y = np.sin(x)

# 2. 创建画布
plt.figure(figsize=(20, 8), dpi=100)
# 3. 绘图
plt.plot(x, y)

plt.grid()

# 4. 显示
plt.show()
```
