# 生成数组的方法

## 生成0和1的数组

```python
# 根据形状生成全1的数组
ones = np.ones((2, 3,4))
ones
```

```python
# 根据已有数组形状生成全0的数组
zeros = np.zeros_like(ones, dtype=np.int64)
zeros
```

## 生成固定范围的数组

```python
# 生成等间隔的数组
a = np.linspace(start=0, stop=100, num=11, endpoint=True)
a
```

```python
# 类似于python的range
b = np.arange(0, 100, 10, dtype=np.int64)
b
```

```python
# 从10^start ~ 10^end 生成等比数列
c = np.logspace(start=0, stop=5, num=6)
c
```

## 均匀分布

```python
# 生成均匀分布的数组
x = np.random.uniform(low=-1, high=1, size=10000000)
x
```

```python
import matplotlib.pyplot as plt
# 1. 创建画布
plt.figure(figsize=(20, 10), dpi=100)
# 2. 绘制直方图
plt.hist(x, bins=1000)

# 3. 显示
plt.show()
```

## 正态分布

```python
# 1. 生成符合正态分布的数
x2 = np.random.normal(loc=2, scale=1, size=10000000)

# 2. 创建画布
plt.figure(figsize=(20, 10), dpi=100)
# 3. 绘制直方图
plt.hist(x2, bins=1000)

# 4. 显示
plt.show()

```

# ndarray切片与索引

```python
# # 案例：随机生成8只股票2周的交易日涨幅数据
stock_change = np.random.normal(loc=0, scale=1, size=(8, 10))
stock_change
```

```python
# 获取第一支股票前三天涨跌幅
stock_change[0, 0:3]
```

```python
# 获取前4只股票, 后4天的涨跌幅
stock_change[0:4, -4:]
# 总结: 切片使用中括号, 每个维度切片都使用逗号隔开
```

```python
# 获取第5只股票第5天的涨跌幅
stock_change[4, 4]
```
