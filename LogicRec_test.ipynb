```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### sigmoid函数

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
#准备数据
nums = np.arange(-10,10,0.1)

fig,ax = plt.subplots(figsize=(20,8))
ax.plot(nums,sigmoid(nums),'b')
# set_xticks(size=25)
plt.tick_params(labelsize=20)       # NOTE: 不是fontsize， 是labelsize
```
