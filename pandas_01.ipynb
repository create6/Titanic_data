```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1,生成10只股票5天的涨跌幅(正态)
stock_change = np.random.normal(0,1,(10,5))
#生成行索引
index = ['股票{}'.format(i) for i in range(stock_change.shape[0])]
#生成列索引,freq='B'跳过周末
date =pd.date_range(start='2019-08-02',periods=stock_change.shape[1],freq='B')


plt.figure(figsize=(20,8),dpi =100)
plt.plot(index,stock_change)
```

```python
#使用数据创建DataFrame
df = pd.DataFrame(stock_change,index=index,columns=date)
df
```
