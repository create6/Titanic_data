# 5. 请简单说明直方图的绘制原理。

```python
# 1、拿到一列数据，保存在列表中；
# 2、求列表内所有元素的最小值min和最大值max；
# 3、建立直角坐标系，在x轴上，把min到max这个区间平均分成N个小区间，N由用户指定；
# 4、根据元素的值的大小，统计列表里落在每个小区间内的元素的数量；
# 5、在x轴上的每个小区间上绘制柱子，柱子的宽度等于小区间的长度，高度等于落在这个小区间上的元素的数量；
# 6、所有的柱子绘制完成后，就得到了这一列数据的直方图。
```

# 13. 在数组[1, 2, 3, 4, 5]中相邻两个数字中间插入两个0

```python
import numpy as np

# 1. 准备数据
arr = np.array([1, 2, 3, 4, 5])
# 2. 生成形状为中间位置插入两个0后的全0数组
# 数组的间隔数量: 原数组长度-1
# 所以新数组长度: 原数组长度 + (原数组长度-1)*2
# 获取数组的长度
len = arr.shape[0]
# 生成新数组
new_arr = np.zeros(len+(len-1)*2)
# 3. 把原数组以步长为3, 放入到新数组中
new_arr[::3] = arr
# 4. 打印新数组
print(new_arr)
```

# 14. 请将以下数据归一化到0~1之间

```python
# 说明: 归一化公式为: (x-min)/(max-min) , 其中min该特征的最小值, max为该特征的最大值, x为具体的特征值
import numpy as np
# 1.准备数据
data = np.array([
    [90,2,10,40],
    [60,4,15,45],
    [75,3,13,46]
])
# 2. 计算每一列的最大值
max_value = np.max(data, axis=0)
# 3. 计算每一列的最小值
min_value = np.min(data, axis=0)
# 4. 根据公式计算,进行归一化处理
new_data = (data - min_value)/(max_value-min_value)
# 5. 打印归一化后的数据
print(new_data)
```
