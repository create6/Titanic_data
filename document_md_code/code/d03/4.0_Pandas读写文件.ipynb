# 缺失值的处理

## 1. 判断缺失值

```python
# 读取电影数据
movies = pd.read_csv('./data/IMDB-Movie-Data.csv')
```

```python
movies.head()
```

```python
import numpy as np
# 判断缺失值
# 如果字段为缺失就是True, 否则为False
movies.isnull()

# 判断文件是否有缺失值
np.any(movies.isnull())
# 这个值为True, 就说明文件中有缺失值

# 如果字段不是缺失为True 是缺失值为False 
# movies.notnull()
# 判断文件是否有缺失值
np.all(movies.notnull())
# 如果这个值为False, 就表示文件中有缺失值.
```

## 处理缺失值

```python
## 1. 删除
# new_movie = movies.dropna()
# 删除缺失值, 默认按行删除; 只要这一行有缺失就删除
# 按列删除缺失值, 只要这一列有缺失值就删除. 
new_movie = movies.dropna(axis=1)
# new_movie
np.any(new_movie.isnull())
# movies
```

```python
## 2. 填充缺失值
# 使用均值填充缺失值; 默认返回填充的数据
# 如果inplace=True, 会修改原来的数据
# movies['Revenue (Millions)'].fillna(value=movies['Revenue (Millions)'].mean(), inplace=True)
# movies

# 填充整个数据的缺失值
for name in movies.columns:
    # print(name)
    # 如果这一列有缺失值, 就使用均值填充
    if np.any(movies[name].isnull()):
        movies.fillna(value=movies[name].mean(), inplace=True)

np.any(movies.isnull())
```
